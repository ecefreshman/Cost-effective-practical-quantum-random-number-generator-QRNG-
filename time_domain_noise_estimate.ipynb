```python
def cc_calculator(file_name):
    seq = np.loadtxt(file_name, dtype=np.float64, delimiter=',')[:5000, 1]
    x = seq[:4900]
    x_2 = x ** 2
    cc = np.zeros(11)
    
    for m in range(11):
        y = seq[m: m + 4900]
        xy = x * y
        y_2 = y ** 2
        cc[m] = (xy.mean() - x.mean() * y.mean()) / np.sqrt((x_2.mean()-x.mean()**2) * (y_2.mean()-y.mean()**2))
    return cc
```

```python
# Sample rate of CC samples is supposed to be lower than the bandwidth.
ass = cc_calculator('test_1_11/50khz.csv')
print(ass)
# ass1 = cc_calculator('test_1_11/1Mhz.csv')
# print(ass1)
# ass2 = cc_calculator('test_1_11/10Mhz.csv')
# print(ass2)
```
