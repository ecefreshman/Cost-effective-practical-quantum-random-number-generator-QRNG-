```python
import numpy as np
# version: numpy 2.3.4
import matplotlib.pyplot as plt
# version: matplotlib 3.9.3
```

```python
def presudo_SA(fft, BW=120, RBW=1):
    '''
    Process raw fft data and pretend I have a SA for a better view.
    :param fft: 
    :param BW: Bandwidth in MHz
    :param RBW: resolution bandwidth in MHz
    :return: 
    '''
    # RBW limit
    RBWL = fft[1, 0] # 6250 Hz in default
    # print('RBWL', RBWL)
    interval = int(RBW * 1e6 / RBWL)
    # print('interval', interval)
    sp = np.zeros(int(BW/RBW) + 1)
    
    start = 1
    stop = int(interval / 2 + 1)
    
    for i in range(int(BW/RBW) + 1):
        sp[i] = np.log10(sum(np.power(fft[start:stop, 1], 2))) * 10
        if i == 0:
            sp[i] = sp[i] + 3
        # print(f'start: {start}, f: {fft[start, 0]}')
        # print(f'stop: {stop}, f: {fft[stop, 0]}')
        start = stop
        stop += interval
    return sp
```

```python
V_e = np.loadtxt("test_1_22/ele_noise.csv", dtype=np.float64, delimiter=",")
V_o = np.loadtxt("test_1_22/osc_noise.csv", dtype=np.float64, delimiter=",")
V_q18 = np.loadtxt("test_1_22/18ma.csv", dtype=np.float64, delimiter=",")
V_q14 = np.loadtxt("test_1_22/14ma.csv", dtype=np.float64, delimiter=",")
V_q10 = np.loadtxt("test_1_22/10ma.csv", dtype=np.float64, delimiter=",")
V_q8 = np.loadtxt("test_1_22/8ma.csv", dtype=np.float64, delimiter=",")
V_q7 = np.loadtxt("test_1_22/7ma.csv", dtype=np.float64, delimiter=",")

freq = np.linspace(0, 120, 121)
sp_q18 = presudo_SA(V_q18)
sp_q14 = presudo_SA(V_q14)
sp_q10 = presudo_SA(V_q10)
sp_q8 = presudo_SA(V_q8)
sp_q7 = presudo_SA(V_q7)
sp_e = presudo_SA(V_e)
sp_o = presudo_SA(V_o)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(freq, sp_q18, label="18ma")
ax.plot(freq, sp_q14, label="14ma")
ax.plot(freq, sp_q10, label="10ma")
ax.plot(freq, sp_q8, label="8ma")
ax.plot(freq, sp_q7, label="7ma")
ax.plot(freq, sp_e, label="ele noise")
ax.plot(freq, sp_o, label="osc noise")
ax.set_xlabel("freqency / MHz")
ax.set_xticks(np.linspace(0, 120, 13))
ax.set_ylabel("$V^2$ / dB")
ax.set_yticks(np.linspace(-92, -74, 19))
ax.grid()
ax.legend()
fig.tight_layout()
plt.show()
```
